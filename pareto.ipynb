```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import mysql.connector
import seaborn as sns
```

```python
conexion = mysql.connector.connect( host='localhost', database ='crepas', user = 'root',password ='Nanda')

# Ejecuta la consulta y almacena los resultados en una variable
results = "SELECT crepa, COUNT(crepa) AS frecuencia, AVG(costo) AS costo_promedio, AVG(precio) AS precio_promedio FROM ventas GROUP BY crepa ORDER BY frecuencia DESC"
df_crepas = pd.read_sql_query(results, con=conexion)
print(df_crepas)
```

```python
plt.figure(figsize=(10,10))
plt.bar(data=df_crepas,x='crepa',height='frecuencia',width = 0.9, color = ['#355070','#6d597a','#b56576','#e56b6f','#eaac8b'])
plt.xticks(rotation = 45)
plt.title('Crepas vendidas')
plt.xlabel('Crepa')
plt.ylabel('Frecuencia')
plt.savefig("Crepas_vendidas")
plt.show()
```
