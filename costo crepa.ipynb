# Costo de cada crepa

## Costo de la mezcla

La receta que se presenta a continuación es para un máximo de hasta 24 crepas. (Según google)$^{1}$

**Ingredientes**

- 2 tazas de harina $\approx$ 400gr
- 2 cucharadas de azúcar $\approx$ 30gr
- 6 huevos
- 1 cucharadita de vainilla $\approx$ 8gr
- 3 tazas de leche $\approx$ 600gr
- 4 cucharadas de mantequilla $\approx$ 60gr
- $\frac{1}{2}$ taza de harina para hot cakes $\approx$ 100gr

De acuerdo con esta receta 1 crepa se hace con:

|**Ingrediente**|**Gramaje**$^{2}$|**Precio**|
|---------------|-----------------|----------|
|Harina|16.7 gr|\$0.32|
|Azúcar|1.25 gr|\$0.06|
|Huevos|0.25 huevos $\approx$ 15.625 gr|\$0.8|
|Vainilla|0.33 gr|\$0.35|
|Leche|25 gr|\$0.53|
|Mantequilla|2.5 gr|\$0.55|
|Harina hot cakes|4.16 gr|\$0.15|
|**Total**|**40.57 gr** |**\$2.76**|
